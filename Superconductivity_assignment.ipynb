```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import curve_fit
```

```python
x = [3.65, 4.5, 5.15, 5.75, 6.3, 6.85, 7.1]
y = [60, 50, 40, 30, 20, 10, 0]
```

```python
def super_curve(T, Bc0, Tc0):
    return Bc0 * (1 - (T/Tc0)**2)
```

```python
initial_guess = [85, 7.1]
params, covariance = curve_fit(super_curve, x, y, p0=initial_guess)
```

```python
params
```

```python
T = np.arange(0, 8, 0.1)
plt.scatter(x, y, color='black', label='Data points')
plt.plot(T, super_curve(T, *params), label='Estimate')
plt.xlim(0, 7.2)
plt.ylim(0, 90)
plt.axhline(y=params[0], color='red', linestyle='--', label=f'$B_c(0) = {params[0]:.1f}\ mT$')
plt.xlabel('T ($in K$)')
plt.ylabel('$B_c \ (in\ mT)$')
plt.title('$B_c\ vs\ T$', fontsize=14)
plt.legend()
plt.savefig('Critical_M.png', dpi=300)
```

```python
T = [3.805, 3.752, 3.659]
M = [113.6, 118.8, 123.8]

ln_T = np.log(T)
ln_M = np.log(M)
# ln_T = -alpha * ln_M + K
```

```python
ln_T, ln_M
```

```python
def linear_fit(ln_M, alpha, k):
    return -alpha*ln_M + k
```

```python
initial_guess = [1, -3]
params, covariance = curve_fit(linear_fit, ln_M, ln_T, p0=initial_guess)
```

```python
params
```

```python
font = {'family': 'serif',
        'color':  'black',
        'weight': 'normal',
        'size': 9,
        }
plt.scatter(ln_M, ln_T, color='black', label="Data points")
plt.plot(np.arange(4.7, 4.9, 0.01), linear_fit(np.arange(4.7, 4.9, 0.01), *params), label='Linear fit')
plt.text(4.8, 1.32, r"$ln\ T = -\alpha ln\ M + k$", fontdict=font)
plt.text(4.8, 1.315, r"$\alpha = 0.453,\ k = 3.48$", fontdict=font)
plt.xlabel('$ln\ M$')
plt.ylabel('$ln\ T$')
plt.xticks(ln_M)
plt.yticks(ln_T)
plt.title('$ln\ T\ vs\ ln\ M$', fontsize=14)
plt.legend()
plt.savefig('isotopic.png', dpi=300)
```
